```python
#Extract the years from the columns
years= df.columns[1:]
total_values = df[years].sum()
total_values
```

```python

plt.figure(figsize=(10,10))
plt.barh(years,total_values,color='blue')
plt.xlabel('Total Values',size=20)
plt.ylabel('Years',size=20)
plt.title('Total Values per year',size=20)
plt.show()
```
